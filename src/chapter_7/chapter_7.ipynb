# Chapter 7: Comparative Analysis

This notebook visualizes the compiled and extrapolated results from the decoders benchmarked in Chapter 6.
It displays the data tables (thresholds, mitigation capacity, error scaling, time scaling, and extrapolations) and the generated plots.

````python
import os
import pandas as pd
from IPython.display import display, Markdown, HTML

data_dir = "../../data/chapter_7"
fig_dir = "../../figures/chapter_7"

csv_files = [
    "analysis_1_thresholds.csv",
    "analysis_2_mitigation.csv",
    "analysis_3_error_scaling.csv",
    "analysis_4_time_scaling.csv",
    "analysis_5_6_extrapolations.csv"
]

plots = [
    "plot_1_thresholds.png",
    "plot_2_mitigation_d9.png",
    "plot_3_error_scaling.png",
    "plot_4_time_scaling.png",
    "plot_5_required_distances.png",
    "plot_6_required_times.png"
]

all_exist = all(os.path.exists(os.path.join(data_dir, c)) for c in csv_files) and \
            all(os.path.exists(os.path.join(fig_dir, p)) for p in plots)

if not all_exist:
    display(Markdown("**Data or plots missing!**"))
    display(Markdown("To generate them, run the following commands from the terminal in the root directory:"))
    display(Markdown("```bash\npython src/chapter_7/analyzer.py process\npython src/chapter_7/analyzer.py plot\n```"))
    # Uncomment to generate directly from the notebook:
    # !python analyzer.py process
    # !python analyzer.py plot
else:
    display(Markdown("## 1. Summary Tables"))
    for c in csv_files:
        display(Markdown(f"### {c}"))
        df = pd.read_csv(os.path.join(data_dir, c))
        display(df.head(10)) # display up to 10 rows per table
    
    display(Markdown("---"))
    display(Markdown("## 2. Analysis Plots"))
    html = '<div style="display: flex; flex-wrap: wrap; justify-content: center;">'
    for p in plots:
        html += f'<div style="margin: 15px; text-align: center;">'
        html += f'<img src="{fig_dir}/{p}" style="width: 450px; border: 1px solid #ddd;" />'
        html += f'<br/><i>{p}</i></div>'
    html += '</div>'
    display(HTML(html))
````
