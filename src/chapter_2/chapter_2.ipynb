### Torus figure
This generates a torus figure for Chapter 2

```python
import numpy as np
import plotly.graph_objects as go

# Torus parameters
R = 3.0  
r = 1  
line_offset = 0.02 

# Generate the 3D surface
u = np.linspace(0, 2 * np.pi, 150)
v = np.linspace(0, 2 * np.pi, 150)
U, V = np.meshgrid(u, v)

X = (R + r * np.cos(V)) * np.cos(U)
Y = (R + r * np.cos(V)) * np.sin(U)
Z = r * np.sin(V)

solid_color = np.zeros_like(Z)

fig = go.Figure(data=[go.Surface(
    x=X, y=Y, z=Z, 
    surfacecolor=solid_color,
    colorscale=[[0, 'rgb(220,220,220)'], [1, 'rgb(220,220,220)']], 
    opacity=1.0, 
    showscale=False
)])

theta = np.linspace(0, 2 * np.pi, 100)
r_line = r + line_offset

# ==========================================
# QUBIT 1
# ==========================================
# Z_L1 (Red): Outer Equator
fig.add_trace(go.Scatter3d(
    x=(R + r_line) * np.cos(theta), y=(R + r_line) * np.sin(theta), z=0 * theta,
    mode='lines', line=dict(color='rgba(255, 0, 0, 0.4)', width=15), showlegend=False
))
# Label Z_L1
fig.add_trace(go.Scatter3d(
    x=[r+0.6], y=[-(R + r + 0.75)], z=[0],
    mode='text', text=['<i>Z<sub>L1</sub></i>'],
    textfont=dict(family="Latin Modern Roman, Times New Roman, serif",color='red', size=52), showlegend=False
))

# X_L1 (Blue): Right vertical loop
fig.add_trace(go.Scatter3d(
    x=R + r_line * np.cos(theta), y=0 * theta, z=r_line * np.sin(theta),
    mode='lines', line=dict(color='rgba(0, 0, 255, 0.4)', width=15), showlegend=False
))
# Label X_L1
fig.add_trace(go.Scatter3d(
    x=[R + 0.5], y=[0], z=[r + 0.35],
    mode='text', text=['<i>X<sub>L1</sub></i>'],
    textfont=dict(family="Latin Modern Roman, Times New Roman, serif",color='blue', size=52), showlegend=False
))

# ==========================================
# QUBIT 2
# ==========================================
# X_L2 (Blue): Inner Equator
fig.add_trace(go.Scatter3d(
    x=(R - r_line) * np.cos(theta), y=(R - r_line) * np.sin(theta), z=0 * theta,
    mode='lines', line=dict(color='rgba(0, 0, 255, 0.4)', width=15, dash='dash'), showlegend=False
))
# Label X_L2
fig.add_trace(go.Scatter3d(
    x=[0], y=[-(R - r - 1.8)], z=[0],
    mode='text', text=['<i>X<sub>L2</sub></i>'],
    textfont=dict(family="Latin Modern Roman, Times New Roman, serif",color='blue', size=52), showlegend=False
))

# Z_L2 (Red): Left vertical loop
fig.add_trace(go.Scatter3d(
    x=-(R + r_line * np.cos(theta)), y=0 * theta, z=r_line * np.sin(theta),
    mode='lines', line=dict(color='rgba(255, 0, 0, 0.4)', width=15, dash='dash'), showlegend=False
))
# Label Z_L2
fig.add_trace(go.Scatter3d(
    x=[-(R + 0.4)], y=[0], z=[r + 0.3],
    mode='text', text=['<i>Z<sub>L2</sub></i>'],
    textfont=dict(family="Latin Modern Roman, Times New Roman, serif",color='red', size=52), showlegend=False
))

# ==========================================
# Display & Export settings
# ========================================== 
camera = dict(
    up=dict(x=0, y=0, z=1),
    center=dict(x=0, y=0, z=0),
    eye=dict(x=0, y=-1.5, z=1)
)

fig.update_layout(
    scene=dict(
        xaxis=dict(visible=False), 
        yaxis=dict(visible=False),
        zaxis=dict(visible=False),
        aspectmode='data'          
    ),
    scene_camera=camera,
    margin=dict(l=0, r=0, b=0, t=0), 
    showlegend=False
)

fig.show()

fig.write_image('../../figures/chapter_2/c2_torus_kitaev.png', width=800, height=800, scale=3)
```
