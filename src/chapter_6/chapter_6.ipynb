### Simulation 1: generating and visualizing the color code
The code below generates a triangular color code with cirq for any given distance and number of rounds. It includes in the code all the necessary detectors and observables for stim to simulate it.

```python
%load_ext autoreload
%autoreload 2
```

```python
import os
import sys
import matplotlib.pyplot as plt
import cirq.contrib.svg
from IPython.display import display, Markdown, SVG


from s1_color_code import ColorCode

print("Generating 2D Color Code lattices and circuits...")

os.makedirs("../../figures", exist_ok=True)

for d in [3, 5, 7, 9]:
    display(Markdown(f"### Distance $d={d}$"))
    
    # Generate the lattice and circuit
    cc = ColorCode(distance=d, lattice_type="4.8.8")
    circ = cc.build_clean_circuit(rounds=1)
    
    # Plot and save lattice as PDF
    lattice_pdf_path = f"../../figures/chapter_6/lattice_d{d}.pdf"
    cc.plot_lattice(save_path=lattice_pdf_path)
    
    # Save circuit as SVG
    svg_content = cirq.contrib.svg.circuit_to_svg(circ)
    svg_content = svg_content.replace('<svg', '<svg style="background-color: white;"')
    
    circuit_svg_path = f"../../figures/chapter_6/circuit_d{d}.svg"
    with open(circuit_svg_path, "w") as f:
        f.write(svg_content)
        
    print(f"Circuit saved to {circuit_svg_path}")
    
    # Display the SVG
    display(SVG(filename=circuit_svg_path))

print("\nGeneration complete.")

```
